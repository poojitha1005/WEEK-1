```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'],bins=30,kde=True)
plt.title('distribution of generated power(kW)')
plt.xlabel('generated power (kW)')
plt.ylabel('frequency')
plt.show()
```

```python
#histogram remaining columns
df[df.columns[:9]].hist(bins=30, figsize=(15,10))
#columns[:9] gives graphs for 9 columns
```

```python
#histograms for remaining columns from 9 to 18
df[df.columns[9:18]].hist(bins=30,figsize=(15,10))
```

```python
#histograms for 18 to 21
df[df.columns[18:21]].hist(bins=30,figsize=(15,10))
```

```python
#bivarite analysis
#scatter plot with target feature
plt.figure(figsize=(20,40))
for i, column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    plt.scatter(df[column],df['generated_power_kw'])
    plt.title(f'{column} vs generated_power(kw)')
    plt.xlabel(column)
    plt.ylabel('generated_power(kw)')
plt.tight_layout
plt.show
```

```python
#corelation
df.corr()
```

```python
plt.figure(figsize=(15,30))
sns.heatmap(df.corr(),cmap='coolwarm',annot=True)
```

```python
#outlier
plt.figure(figsize=(15,30))
for i,column in enumerate(df.columns):
    plt.subplot(7,3,i+1)
    sns.boxplot(df[column])
plt.show()
```
